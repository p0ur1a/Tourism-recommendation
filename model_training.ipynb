```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
import seaborn as sns
```

```python
df= pd.read_csv("model_training_dataset.csv")
```

```python
# Data Preprocessing
df['Region'] = df['Region'].str.lower()

# Splitting data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(df['Region'], df['Region-Label'], test_size=0.2, random_state=42)

# Converting text to numerical features using TfidfVectorizer
vectorizer1 = TfidfVectorizer(stop_words='english', max_features=5000)
X_train_vec = vectorizer1.fit_transform(X_train)
X_test_vec = vectorizer1.transform(X_test)

# Building and training the model
model1 = MultinomialNB()
model1.fit(X_train_vec, y_train)

# Predicting using the model
y_pred = model1.predict(X_test_vec)

# Evaluating the model
region_accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f'Accuracy: {region_accuracy}')
print('Classification Report:')
print(report)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score, classification_report

# Data Preprocessing
df['Season'] = df['Season'].str.lower()

# Splitting the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(df['Season'], df['Season-Label'], test_size=0.2, random_state=42)

# Converting text to numerical features using TfidfVectorizer
vectorizer2 = TfidfVectorizer(stop_words='english', max_features=5000)
X_train_vec = vectorizer2.fit_transform(X_train)
X_test_vec = vectorizer2.transform(X_test)

# Building and training the SVM model
model2 = SVC(kernel='linear', random_state=42)
model2.fit(X_train_vec, y_train)

# Making predictions using the SVM model
svm_y_pred = model2.predict(X_test_vec)

# Evaluating the SVM model
season_accuracy = accuracy_score(y_test, svm_y_pred)
svm_report = classification_report(y_test, svm_y_pred)

print(f'SVM Accuracy: {season_accuracy}')
print('SVM Classification Report:')
print(svm_report)
```

```python
# Preprocess the data
df['Accommodation'] = df['Accommodation'].str.lower()

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['Accommodation'], df['Accommodation-Label'], test_size=0.2, random_state=42)

# Convert text data to TF-IDF features
vectorizer3 = TfidfVectorizer(stop_words='english', max_features=5000)
X_train_vec = vectorizer3.fit_transform(X_train)
X_test_vec = vectorizer3.transform(X_test)

# Build and train the Random Forest model
model3 = RandomForestClassifier(n_estimators=100, random_state=42)
model3.fit(X_train_vec, y_train)

# Make predictions with the Random Forest model
rf_y_pred = model3.predict(X_test_vec)

# Evaluate the Random Forest model
accommodation_accuracy = accuracy_score(y_test, rf_y_pred)
rf_report = classification_report(y_test, rf_y_pred)

accommodation_accuracy, rf_report
print(f'accommodation_accuracy: {accommodation_accuracy}')
```

```python
#Applying clustering to Budget column
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

# Extract the text column
text_data = df['Budgets']

# Vectorize the text using TF-IDF
vectorizer6 = TfidfVectorizer(stop_words='english')
X = vectorizer6.fit_transform(text_data)

# Determine the number of clusters using the Elbow Method
inertia = []
range_clusters = range(1, 11)

for n_clusters in range_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Plot the Elbow chart
plt.figure(figsize=(8, 6))
plt.plot(range_clusters, inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Determining the Optimal Number of Clusters')
plt.show()

# Choose the optimal number of clusters and train the KMeans model
n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(X)

# Assign clusters to each data point
df['Cluster'] = kmeans.labels_

# Display some examples from each cluster
cluster_examples = df.groupby('Cluster').head(3)
print(cluster_examples)

# Silhouette Score
silhouette_avg = silhouette_score(X, kmeans.labels_)
print(f'Silhouette Score: {silhouette_avg}')

# Visualization of clusters
plt.figure(figsize=(10, 6))
colors = ['red', 'blue', 'green']
for i in range(n_clusters):
    cluster_data = df[df['Cluster'] == i]
    plt.scatter(cluster_data.index, [i]*len(cluster_data), c=colors[i], label=f'Cluster {i}')

plt.xlabel('Index')
plt.ylabel('Cluster')
plt.title(f'Cluster Visualization (Silhouette Score: {silhouette_avg:.2f})')
plt.legend()
plt.show()
```
